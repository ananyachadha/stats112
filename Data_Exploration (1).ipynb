```python
import pandas as pd
data = pd.read_csv("energyandweather.csv")
dataclean = data.iloc[10:,:]
data.columns.values
dataclean.to_csv("dataclean.csv")
```

```python
import plotly as plt
import plotly.express as px
data["YearAndMonth"] =  data["DateAPIFormat"].astype(str).str[:-2]
averages = data.groupby("YearAndMonth")["Return $/MW"].mean()
averages.sort_index().plot.bar(title="Average Price Spread (%) Per Month", ylabel="Price Spread")
```

no obvious relationship with month

```python
data["YearAndMonth"] =  data["DateAPIFormat"].astype(str).str[:-2]
averages = data.groupby("YearAndMonth")["RT Heat Rate "].mean()
averages.sort_index().plot.bar(title="Average Heat Rate Per Month", ylabel="Price Spread")
```
